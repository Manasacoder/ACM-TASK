```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv(r'D:\Manasa\top5-players.csv')
  

# Data cleaning and preprocessing

print(data.isnull().sum())  

data_cleaned= data.fillna(0)                                                  #Handling Missing Values                              
data_cleaned = data_cleaned.drop_duplicates()                                 #Remove Duplicate Entries
#df_cleaned.head()

# Data Analysis

forwards = data_cleaned[data_cleaned['Pos'].str.contains('FW')]
midfielders = data_cleaned[data_cleaned['Pos'].str.contains('MF')]
defenders = data_cleaned[data_cleaned['Pos'].str.contains('DF')]


                                                 #   strikers by goals + assists per 90 minutes (G+A_90)
top_forwards = forwards.sort_values(by='G+A_90', ascending=False).head(10)
top_forward = forwards.loc[forwards['G+A_90'].idxmax()]

plt.figure(figsize=(10,6))
sns.barplot(x='Player', y='G+A_90', data=top_forwards)
plt.xticks(rotation=90)                                                                 
plt.title('Top 10 forwards Based on Goals+Assists per 90 mins')
plt.show()


                                                  #  midfielders by assists per 90 mins (Ast_90)
top_midfielders = midfielders.sort_values(by='Ast_90', ascending=False).head(10)
top_midfielder = midfielders.loc[midfielders['Ast_90'].idxmax()]

plt.figure(figsize=(10,6))
sns.barplot(x='Player', y='Ast_90', data=top_midfielders)
plt.xticks(rotation=90)
plt.title('Top 10 Midfielders Based on Assists per 90 mins')
plt.show()

                                                 # defenders by tackles per 90 mins (Tkl_90)
top_defenders = defenders.sort_values(by='PrgP', ascending=False).head(10)
top_defender = defenders.loc[defenders['PrgP'].idxmax()]

plt.figure(figsize=(10,6))
sns.barplot(x='Player', y='PrgP', data=top_defenders)
plt.xticks(rotation=90)
plt.title('Top 10 Defenders Based on Progressive Passes')
plt.show()



# summarizing key insights.

print("Top Players by Position:")
print(f"Top Forward: {top_forward['Player']} (G+A_90: {top_forward['G+A_90']})")
print(f"Top Midfielder: {top_midfielder['Player']} (Ast_90: {top_midfielder['Ast_90']})")
print(f"Top Defender: {top_defender['Player']} (PrgP: {top_defender['PrgP']})")


```
